```python
import pandas as pd
import json

#Read in smaller file of Yelp Business data
df = pd.read_json('Business5000.json')
df.head(30)
```

```python
# Verify number of records and columns read in
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# Get all the state values for the dataframe
df.state.value_counts()
```

```python
# Drop the 'attributes', 'hours', latitude', & 'longitude' columns
df.drop(['attributes', 'latitude', 'longitude', 'hours'], inplace=True, axis=1)
df.head()
```

```python
# Keep only rows with 'is_open' == 1
open_df = df.loc[df['is_open'] == 1]
open_df.head()
```

```python
open_df.is_open.value_counts()
```

```python
# Keep only rows with "Restaurants" appearing anywhere in the 'categories' column
Rest_df = open_df[open_df["categories"].str.contains("Restaurants")==True]
Rest_df.head()
```

```python
Rest_df.shape
```

```python
# Drop the 'is_open' and 'categories' columns
df.drop(['is_open', 'categories'], inplace=True, axis=1)
df.head()
```

```python
Rest_df.dtypes
```

```python
df.to_csv(r'Cleaned_Business.csv')
```
